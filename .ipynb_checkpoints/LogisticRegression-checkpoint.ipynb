```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in csv file
train = pd.read_csv("records.csv")
train.head()
```

```python
X = train.drop(columns=["User Name","Friend Name","Friendship Status"],axis=1)
X.head()
```

```python
Y = train.iloc[:,12:13]
Y.head()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e5, solver='lbfgs',class_weight='balanced')
logreg.fit(X, Y.values.ravel())
```

```python
z = logreg.predict(X)
h = logreg.decision_function(X)
# max = 0
# for i in h:
#     print(i)
#         max = i
# print(max)
```
